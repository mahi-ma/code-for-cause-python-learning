```python
# Question 3

def gcd(a,b):
    if (a == 0):
        return b
    elif (b == 0):
        return a
    else:
      if (a == b):
        return a
 
      if (a > b):
        return gcd(a-b, b)
      else:
        return gcd(a, b-a)

first = int(input())
second = int(input())

print("GCD = ", gcd(first,second))
```
