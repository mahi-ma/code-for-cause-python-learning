```python
# Question 1
def decimal_to_octal(num):
  octal = 0
  i = 1
  while (num != 0):
      octal = octal + (num % 8) * i
      num = int(num / 8)
      i = i * 10;

  return octal;

decimal = int(input())
print(decimal_to_octal(decimal))
```

```python
# Question 2
n1 = int(input())
n2 = int(input())
ch = input()

if ch=="*":
  print(n1*n2)
elif ch=="/":
  print(n1/n2)
elif ch=="%":
  print(n1%n2)
elif ch=="+":
  print(n1+n2)
elif ch=="-":
  print(n1 - n2)
else:
  print("Please enter ch as one of *,/,%,+,-")
```

```python
# Question 7

a = int(input())
b = int(input())

def LCM(a,b):
  if a > b:
    bigNo = a
  else:
    bigNo = b
  while(True):
      if((bigNo % a == 0) and (bigNo % b == 0)):
          lcm = bigNo
          break
      bigNo += 1

  return lcm

print("LCM = ",LCM(a,b))

```
