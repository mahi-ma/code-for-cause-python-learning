```python
# Question 1
def decimal_to_octal(num):
  octal = 0
  i = 1
  while (num != 0):
      octal = octal + (num % 8) * i
      num = int(num / 8)
      i = i * 10;

  return octal;

decimal = int(input())
print(decimal_to_octal(decimal))
```

```python
# Question 7

a = int(input())
b = int(input())

def LCM(a,b):
  if a > b:
    bigNo = a
  else:
    bigNo = b
  while(True):
      if((bigNo % a == 0) and (bigNo % b == 0)):
          lcm = bigNo
          break
      bigNo += 1

  return lcm

print("LCM = ",LCM(a,b))

```

```python
# Question 9

def isPrime(val):
  prime = True
  for i in range(2,int(val / 2)):
    if(val % i == 0 and i!=val):
      prime = False
      break
  return prime

n = int(input())  #number of elements in array
count = 0
for i in range(n):
  p = int(input())
  if(isPrime(p)):
    count+=1
  else:
    continue

print("number of elements which are prime are ", count)
```

```python
# Question 12
# we can use similar thing like sort 0's and 1's which is dutch national flag algo but that would segregate it out of order, so did this to segregate in order in O(n2)

arr = [int(item) for item in input("Enter the list items : ").split()]

for i in range(int(len(arr) / 2)):
  for j in range(len(arr) - 1):
    if (arr[j] % 2 !=0) and (arr[j+1] % 2 ==0):
      temp = arr[j]
      arr[j] = arr[j+1]
      arr[j+1] = temp

print(arr)
```

```python
# Question 14

inputArr = [int(item) for item in input("Enter List : ").split()]
target = int(input("Enter target : "))
output = []

for i in range(len(inputArr)):
  complement = target - inputArr[i]
  if (complement in inputArr[i+1:]) and ((complement,inputArr[i]) not in output):
    output.append((inputArr[i],complement))

print(output)
    
```
